```python
from pathlib import Path
import pandas as pd

problem_files = [
    ("Jongkok", "Bustan", f"{i}.Csv") for i in range(1, 10)
] + [
    ("Jatuh", "Bustan", f"{i}.Csv") for i in range(6, 10)
]

rows = []

for activity, subject, raw_file in problem_files:
    clean_file = raw_file.replace(".Csv", ".csv")
    
    raw_path = Path("/media/spell/Spell-lab/Lidar/Dataset") / activity / subject / "CSV" / raw_file
    clean_path = Path("/media/spell/Spell-lab/Lidar/Clean Dataset") / activity / subject / clean_file
    
    rec = {
        "activity": activity,
        "subject": subject,
        "file": clean_file,
        "raw_exists": raw_path.exists(),
        "raw_size_bytes": raw_path.stat().st_size if raw_path.exists() else None,
        "raw_rows": None,
        "raw_cols": None,
        "clean_exists": clean_path.exists(),
        "clean_size_bytes": clean_path.stat().st_size if clean_path.exists() else None,
        "clean_rows": None,
        "clean_cols": None,
        "raw_error": None,
        "clean_error": None,
    }
    
    try:
        if raw_path.exists():
            raw_df = pd.read_csv(raw_path)
            rec["raw_rows"] = len(raw_df)
            rec["raw_cols"] = len(raw_df.columns)
    except Exception as e:
        rec["raw_error"] = str(e)
    
    try:
        if clean_path.exists():
            clean_df = pd.read_csv(clean_path)
            rec["clean_rows"] = len(clean_df)
            rec["clean_cols"] = len(clean_df.columns)
    except Exception as e:
        rec["clean_error"] = str(e)
    
    rows.append(rec)

empty_diagnosis_df = pd.DataFrame(rows)
display(empty_diagnosis_df)
```

```python
from pathlib import Path
import pandas as pd

RAW_CANDIDATES = [
    Path("/media/spell/Spell-lab/Lidar/Dataset"),
    Path("/media/spell/Spell-lab/Lidar/Raw Dataset"),
]

problem_files = [
    ("Jongkok", "Bustan", f"{i}.Csv") for i in range(1, 10)
] + [
    ("Jatuh", "Bustan", f"{i}.Csv") for i in range(6, 10)
]

rows = []

for activity, subject, raw_file in problem_files:
    clean_file = raw_file.replace(".Csv", ".csv")
    clean_path = Path("/media/spell/Spell-lab/Lidar/Clean Dataset") / activity / subject / clean_file
    
    rec = {
        "activity": activity,
        "subject": subject,
        "file": clean_file,
        "clean_exists": clean_path.exists(),
        "clean_size_bytes": clean_path.stat().st_size if clean_path.exists() else None,
        "clean_rows": None,
    }
    
    try:
        if clean_path.exists():
            clean_df = pd.read_csv(clean_path)
            rec["clean_rows"] = len(clean_df)
    except Exception as e:
        rec["clean_error"] = str(e)
    
    for raw_base in RAW_CANDIDATES:
        raw_path = raw_base / activity / subject / "CSV" / raw_file
        key_prefix = raw_base.name.replace(" ", "_").lower()
        
        rec[f"{key_prefix}_exists"] = raw_path.exists()
        rec[f"{key_prefix}_size_bytes"] = raw_path.stat().st_size if raw_path.exists() else None
        rec[f"{key_prefix}_rows"] = None
        rec[f"{key_prefix}_cols"] = None
        rec[f"{key_prefix}_error"] = None
        
        try:
            if raw_path.exists():
                raw_df = pd.read_csv(raw_path)
                rec[f"{key_prefix}_rows"] = len(raw_df)
                rec[f"{key_prefix}_cols"] = len(raw_df.columns)
        except Exception as e:
            rec[f"{key_prefix}_error"] = str(e)
    
    rows.append(rec)

raw_location_check_df = pd.DataFrame(rows)
display(raw_location_check_df)
```
